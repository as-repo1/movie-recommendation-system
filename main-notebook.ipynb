```python
movies = movies_raw.merge(credits_raw,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(3)
```

```python
# convert the string to list
def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
# drop duplicates
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    l =[]
    counter=0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
            counter+=1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head(3)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(10)
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(' ',''))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
# movies['production_companies'] = movies['production_companies'].apply(collapse)
# movies['production_countries'] = movies['production_countries'].apply(collapse)
# movies['spoken_languages'] = movies['spoken_languages'].apply(collapse)
```

```python
movies.sample(3)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.sample(5)
# new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
# to retrieve the index of a specific movie in the DataFrame new. It's finding the index corresponding to the movie with the title 'The Dark Knight Rises'. 
new[new['title'] == 'The Dark Knight Rises'].index[0] # 3
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
import gradio as gr

def recommend_movies(movie_title, num_recommendations=5):
    index = new[new['title'] == movie_title].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    # Limit the number of recommendations based on the slider value
    recommended_movies = [new.iloc[i[0]].title for i in distances[1:num_recommendations+1]]
    return "\n".join(recommended_movies)

demo = gr.Interface(
    fn=recommend_movies, 
    inputs=["text", gr.inputs.Slider(1, 13, step=1)],  # Use step=1 to allow only integer values
    outputs="text",  # Use "text" type for displaying recommendations
    live=True
)

demo.launch(share=True)
```
